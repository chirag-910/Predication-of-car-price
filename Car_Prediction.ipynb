```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
car_data = pd.read_csv('CarPrice_Assignment.csv')
car_data.head()
```

```python
car_data.set_index('car_ID',inplace=True)
car_data.head()
```

## Data Pre-Processing

```python
# Extracting Company Name 
car_data['CarName'] = car_data['CarName'].str.split(' ',expand=True)[0].str.lower()
car_data.head()
```

```python
# Checking if nan value is there
car_data[pd.isnull(car_data)].count()
```

```python
car_data['CarName'].value_counts()
```

```python
#Resolving Amiguity in Company Name
car_data['CarName'] = car_data['CarName'].replace({'toyouta':'toyota','vokswagen':'volkswagen','porcshce':'porsche'})
car_data['CarName'].value_counts()
```

## DUMMY VARIABLES

```python
#dummy variable for car name
CarName = pd.get_dummies(car_data['CarName'],prefix='CarName',drop_first=True)
CarName.drop('CarName_mercury',axis=1,inplace=True)
car_data = pd.concat([car_data,CarName],axis=1)

#Dummy variable for fuelGas
gas_fuelType = pd.get_dummies(car_data['fueltype'],prefix='FuelType',drop_first=True)
car_data['gas_fueltype'] = gas_fuelType

#dummy variable for aspiration
std_aspiration = pd.get_dummies(car_data['aspiration'],prefix='aspiration').iloc[:,0:1]
car_data['aspiration_std'] = std_aspiration

#dummy vairable for numberoOfDoor
four_door_number = pd.get_dummies(car_data['doornumber'],prefix='doorNumner').iloc[:,0:1]
car_data['doorNumber_four'] = four_door_number

#Dummy variable for carbody
carbody = pd.get_dummies(car_data['carbody'],prefix='carbody',drop_first=True)
car_data = pd.concat([car_data,carbody],axis=1)

#dummy variable for driveWheel
d_wheel = pd.get_dummies(car_data['drivewheel'],prefix='wheel',drop_first=True)
car_data = pd.concat([car_data,d_wheel],axis=1)

#Dummy variable for engloc
engin_loc = pd.get_dummies(car_data['enginelocation'],prefix='engloc').iloc[:,0:1]
car_data['engloc_front'] = engin_loc

#Dummy variable for eng_type
eng_type = pd.get_dummies(car_data['enginetype'],prefix='engtype')
eng_type.drop('engtype_dohc',axis=1,inplace=True)
car_data = pd.concat([car_data,eng_type],axis=1)

#Dummy variable for cylindernumber
cyc_num = pd.get_dummies(car_data['cylindernumber'],prefix='cycNum')
cyc_num.drop('cycNum_twelve',axis=1,inplace=True)
car_data = pd.concat([car_data,cyc_num],axis=1)

#dummy vairable for fuelsystem
fuel_sys = pd.get_dummies(car_data['fuelsystem'],prefix='fuelSys')
fuel_sys.drop('fuelSys_spfi',axis=1,inplace=True)
car_data = pd.concat([car_data,fuel_sys],axis=1)

#dummy variable for symboling
symb = pd.get_dummies(car_data['symboling'],prefix='symb',drop_first=True)
car_data = pd.concat([car_data,symb],axis=1)
```

### Spliting Data into train test split

```python
from sklearn.model_selection import train_test_split

# We specify this so that the train and test data set always have the same rows, respectively

df_train, df_test = train_test_split(car_data, train_size = 0.7, test_size = 0.3, random_state = 101)
```

#### Rescaling the features 

```python
# USING MINMAX method
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
# below are the column values which have the high numberic value and requires scaling
var = ['price','wheelbase','carlength','carwidth', 'carheight','curbweight',
       'enginesize', 'boreratio', 'stroke', 'compressionratio', 'horsepower',
       'peakrpm', 'citympg', 'highwaympg']
```

```python
# Scaling the function using scaler method
df_train[var] = scaler.fit_transform(df_train[var])
df_train.head()
```

```python
car_data.info()
```

### Dividing into x and y for modeling

```python
y_train = df_train.pop('price')
X_train = df_train
```

### USING RFE to find top 10 feature

```python
# Importing RFE and LinearRegression
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
# Running RFE with the output number of the variable equal to 10
lm = LinearRegression()
lm.fit(X_train, y_train)

rfe = RFE(lm, 10)             # running RFE
rfe = rfe.fit(X_train, y_train)
```

```python
list(zip(X_train,rfe.support_,rfe.ranking_))
```

```python
# top 10 features required for modeling
col = X_train.columns[rfe.support_]
col
```

### Building model using statsmodel, for the detailed statistics

```python
# Creating X_test dataframe with RFE selected variables
X_train_car_data = X_train[col]
```

```python
X_train_car_data.head()
```

```python
# Adding a constant variable 
import statsmodels.api as sm  
X_train_car_data = sm.add_constant(X_train_car_data)
```

```python
lm1 = sm.OLS(y_train,X_train_car_data).fit()   # Running the linear model
```

```python
print(lm1.summary())
```

Here we see that all the features have p-values under accepted range<br>
Also R-Square and Adjusted R-Sqaure are 95.1% and 94.8% respectively<br>
Now we will check the vif value for all features

### Checking VIF 

```python
X_train_car_data.head()
```

```python
#Removing constant
X_train_car_data = X_train_car_data.drop('const',axis=1)
```

```python
# Calculate the VIFs for the new model
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
X = X_train_car_data
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```
